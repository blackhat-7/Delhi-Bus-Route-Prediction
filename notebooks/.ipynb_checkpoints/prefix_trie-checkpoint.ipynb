```python
feed = gtfs_realtime_pb2.FeedMessage()
lat=[]
lon=[]
for i in range(100):
  response = urllib.request.urlopen('https://opendata.iiitd.edu.in/api/realtime/VehiclePositions.pb?key=5kOKha1VJqxWvkkIp1J71Yrb5mc3VsDS')
  feed.ParseFromString(response.read())
  put(feed,lat,lon)
  time.sleep(10)
```
